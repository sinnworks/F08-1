# [dbo].[SelectProducts]

```sql
/****** Object:  StoredProcedure [dbo].[SelectProducts]    Script Date: 2023/9/11 下午 10:43:50 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[SelectProducts]
(
@CategoryId INT=NULL
)
AS
BEGIN;
	SET NOCOUNT ON;

	SELECT P.Id as ProductId, P.Name as ProductName, C.Name as CategoryName
	FROM Products as P
	Inner Join Categories as C on C.Id = P.CategoryId
	WHERE (@CategoryId IS NULL OR P.CategoryId = @CategoryId)
	ORDER BY C.DisplayOrder

	SET NOCOUNT OFF;

END;
GO
```
